Variables Mal46p, Mal66p, Mal82p and Mal94p describe the percentage of 1995 population in malaria-affected areas in four different years (1946, 1966, 1982 and 1994). Generate a variable that categorizes observations in Mal46p (percentage of population) by low, moderate and high, based on the following groupings:
a. Low if less than or equal to 33 percent of the 1995 population resided in a malaria-affected area in 1946.
b. Moderate if more than 33 percent and less than or equal to 66 percent of the 1995 population resided in a malaria-affected area in 1946.
c. High if more than 66 percent of the 1995 population resided in a malaria-affected area in 1946.
Repeat the process and generate three more variables that categorize observations for the three remaining years. Use a loop to accomplish this task.

```python
#Renaming column names for better understanding.
df2 = df2.rename(columns = {'% malaria area in 1946' : 'Mal46p' , 
                            '% malaria area in 1966':'Mal66p', 
                            '% malaria area in 1982':'Mal82p',
                            '% malaria area in 1994' : 'Mal94p'})
df2
                            
```

```python
#categorizing observations


df2.loc[(df2['Mal46p'] <= 0.3),'Type46'] = 'Low'
df2.loc[(df2['Mal46p'] >= 0.3 ) & (df2['Mal46p'] <= 0.6 ),'Type46'] = 'Moderate'
df2.loc[(df2['Mal46p'] >= 0.6 ),'Type46'] = 'High'


df2.loc[(df2['Mal66p'] <= 0.3),'Type66'] = 'Low'
df2.loc[(df2['Mal66p'] >= 0.3 ) & (df2['Mal66p'] <= 0.6 ),'Type66'] = 'Moderate'
df2.loc[(df2['Mal66p'] >= 0.6 ),'Type66'] = 'High'


df2.loc[(df2['Mal82p'] <= 0.3),'Type82'] = 'Low'
df2.loc[(df2['Mal82p'] >= 0.3 ) & (df2['Mal82p'] <= 0.6 ),'Type82'] = 'Moderate'
df2.loc[(df2['Mal82p'] >= 0.6 ),'Type82'] = 'High'


df2.loc[(df2['Mal94p'] <= 0.3),'Type94'] = 'Low'
df2.loc[(df2['Mal94p'] >= 0.3 ) & (df2['Mal94p'] <= 0.6 ),'Type94'] = 'Moderate'
df2.loc[(df2['Mal94p']>= 0.6 ),'Type94'] = 'High'
df2
```

Normalize the variable constructed in (16) using the mean and standard deviation of the dataset which results from (16).

```python
def std_norm(df2, column):
    c = df2[column]
    df2[column] = (c - c.mean())/c.std()
norm = std_norm(df2, 'value')
print(norm)
```
